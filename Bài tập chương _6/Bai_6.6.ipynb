```python
import threading

def print_even_numbers(max_num):
    for num in range(2, max_num + 1, 2):
        print(f"Even: {num}")

def print_odd_numbers(max_num):
    for num in range(1, max_num + 1, 2):
        print(f"Odd: {num}")

def main():
    max_num = 20  # Ngưỡng tối đa

    even_thread = threading.Thread(target=print_even_numbers, args=(max_num,))
    odd_thread = threading.Thread(target=print_odd_numbers, args=(max_num,))

    even_thread.start()
    odd_thread.start()

    even_thread.join()
    odd_thread.join()


if __name__ == "__main__":
    main()
import threading
import time
from datetime import datetime

def task(name):
    print(f"Starting {name}")
    for _ in range(3):
        print(f"{name}: Web {datetime.now().strftime('%a %b %d %H:%M:%S %Y')}")
        time.sleep(1)
    print(f"Exiting {name}")

def main():
    print("Starting Main Thread")

    threads = []
    services = ["Google", "Yahoo", "Facebook"]

    for service in services:
        thread = threading.Thread(target=task, args=(service,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    print("Exiting Main Thread")

if __name__ == "__main__":
    main()
```
